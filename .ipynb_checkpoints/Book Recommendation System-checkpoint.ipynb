# Performing EDA on each Dataset

```python
# Checking total data 
books.shape
ratings.shape
```

```python
# Merge datasets
merged_data = pd.merge(ratings, books, on='ISBN')
merged_data
```

```python
# Explore the distribution of user ratings
sns.histplot(data=merged_data, x='Book-Rating', bins=10)
plt.title('Distribution of User Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

# Creating the Book Recommendation System based on Popularity 

```python
num_ratingdf = merged_data.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratingdf.rename(columns={'Book-Rating':'num_ratings'}, inplace = True)
num_ratingdf
```

```python
avg_ratingdf = merged_data.groupby('Book-Title').count()['Book-Rating'].reset_index()
avg_ratingdf.rename(columns={'Book-Rating':'avg_rating'}, inplace = True)
avg_ratingdf
```

```python
popularity_df = num_ratingdf.merge(avg_ratingdf, on='Book-Title')
```

```python
popular_df = popularity_df[popularity_df['num_ratings']>=150].sort_values('avg_rating',ascending = False)
```

```python
 popular_df
```

```python
popular_df = popular_df.merge(books, on= 'Book-Title').drop_duplicates('Book-Title')[['Book-Title', 'Book-Author','Image-URL-M', 'num_ratings', 'avg_rating']]
```
